```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}


# Поиск владельца документа
def function_people():
    number_input = input('Введите номер документа : ')
    finded=0
    name_doc=''
    for el in documents:
        if el['number'] == number_input:
            finded=1
            name_doc=el['name']
    if finded:
        print('Владелец документа: ',name_doc)
    else:
        print('Документ не найден в базе')

# Поиск полки, на которой расположен документ:
def function_shelf():
    number_directories = 0
    number_input = input('Введите номер документа : ')
    for key, value in directories.items():
        if number_directories == 0:
            for el in value:
                if number_input in value:
                    number_directories = key
    if number_directories !=0:
        print('Документ №', number_input, 'на полке №', number_directories)
    else:
        print('Документа №', number_input, 'нет на полках')

#Вывод списка документов: 
def function_list():
    print('Список документов :')
    for el in documents:
        print(f"\"{el['type']}\" \"{el['number']}\" \"{el['name']}\"")

#Добавление новой полки: 
def function_as():
    
    number_input = input('Введите номер полки : ')
    if number_input in directories:
        print('Полка с таким номером уже существует')
    else:
    #    print('Добавляю полку')
        directories[number_input]=[]
    print(directories)

#Удаление документа:    
def function_delete():
    number_input = input('Введите номер документа : ')
    finded = 0
    for el in documents:
        if el['number'] == number_input:
            documents.remove(el)
            finded = 1
            print('Удаляем документ : ',el['type'],el['number'],el['name'])
    for value in directories.values():
        for el in value:
            if number_input == el:
                value.remove(el)
    if(finded == 0):
        print('Документ c номером',number_input,' не существует')
    
def function_move():
    number_input = input('Введите номер документа : ')
    shelf_input = input('Введите номер полки : ')
    finded_number = 0
    finded_shelf = 0
    for key, value in directories.items():
        for i, el in enumerate(value):
            if finded_number == 0:
                if number_input == el:
                    finded_number=1
                    finded_shelf=key
                    #print('Документ :',el,'был на полке:',finded_shelf)
                    #print('УДАЛЯЮ el=',el)
                    if shelf_input in directories: #нашли полку
                        value.remove(el)
                        print('Перемещаем документ на полку',shelf_input)
                        directories[shelf_input].append(number_input)
                    else:
                         print('Нет полки для перемещения документа, оставляем на месте')                   
    if finded_number == 0: #не нашли документ
         print('Документа №', number_input, 'нет на полках') 
    
def function_add():
    type_input = input('Введите тип документа: ')
    number_input = input('Введите номер документа: ')
    name_input = input('Введите имя владельца: ')
    shelf_input = input('Введите номер полки: ')

    if shelf_input in directories: 
        print('Новый документ перемещен на полку',shelf_input)
        new_document = dict(type=type_input,number=number_input,name=name_input)
        documents.append(new_document)
        directories[shelf_input].append(number_input)
    else:
        print('Введён номер несуществующей полки, документ не будет размещен')

# Справка
def help_text():            
    print("Добро пожаловать! Вам доступны следующие команды:")
    print("p для поиска владельца документа")
    print("s для поиска места распложения документа")
    print("l для вывода списка всех документов")
    print("d для удаления документа")
    print("m для перемещения документа")
    print("a для добавления нового документа")
    print("q для выхода")        
        
# Финальный блок        
def main():
    help_text()
    while True:
        user_input = input('Введите команду: ')
        if user_input == 'p':
            function_people()
        elif user_input == 's':
            function_shelf()
        elif user_input == 'l':
            function_list()
        elif user_input == 'as':
            function_as()
        elif user_input == 'd':
            function_delete()
        elif user_input == 'm':
            function_move()
        elif user_input == 'a':
            function_add()
        elif user_input == 'q':
            break
        else: 
            print('Неизвестная команда!')
            break
        print()
        
main()
```
